```python
data['Ram'] = data['Ram'].str.replace('GB','').astype('int32')  #  to make Ram as int  
data['Weight'] = data['Weight'].str.replace('kg','').astype('float32')  #  to make Ram as intweignt     
```

```python
def add_company(inpt):
    # Check if the company is in the specified list or if its count is less than 10
    if inpt in ['Samsung', 'Razer', 'Mediacom', 'Microsoft', 'Xiaomi', 'Vero', 'Huawei','Chuwi','Google','Fujitsu','LG']:
        return 'Other'
    else:
        return inpt

# Apply the function to the 'Company' column
data['Company'] = data['Company'].apply(add_company)
```

```python
data['Touchscreen'] = data['ScreenResolution'].apply(lambda x:1 if 'Touchscreen' in x else 0)
data['Ips'] = data['ScreenResolution'].apply(lambda x:1 if 'IPS' in x else 0) # created new column named Touchscree
```

```python
data['Cpu_name'] = data['Cpu'].apply(lambda x:" ".join(x.split()[0:3]))
```

```python
def set_processor(name):
    if name == 'Intel Core i7' or name == 'Intel Core i5' or name == 'Intel Core i3':
        return name
    elif name.split()[0] == 'AMD':
        return 'AMD'
    else:
        return 'Other'

# Apply the function to the 'cpu_name' column
data['Cpu_name'] = data['Cpu_name'].apply(set_processor)
```

```python
data = data[data['Gpu_name'] != 'ARM'] # remove ARM becouse it has only 1
```

```python
data['Touchscreen'] = data['ScreenResolution'].apply(lambda x:1 if 'Touchscreen' in x else 0)
data['Ips'] = data['ScreenResolution'].apply(lambda x:1 if 'IPS' in x else 0) # created new column named Touchscreen and lps
```

```python
data.head()
```

```python
data['OpSys'].value_counts()
```

```python
def set_os(inpt):
    if inpt == 'Windows 10' or inpt == 'Windows 7' or inpt == 'Windows 10 S':
        return 'Windows'
    elif inpt == 'macOS' or inpt == 'Mac OS X':
        return 'Mac'
    elif inpt == 'Linux':
        return 'Linux'
    else:
        return 'Other'

# Apply the function to the 'OpSys' column
data['OpSys'] = data['OpSys'].apply(set_os)
```

```python
data = data.drop(columns=['laptop_ID', 'Inches', 'Product', 'ScreenResolution', 'Cpu', 'Gpu'])
```

```python
data.head(5)
```

```python
data.info()
```

```python
data = pd.get_dummies(data) #one hot encoding                    
  
```

```python
data.head(2)
```

```python
# List of boolean columns to convert (excluding 'Weight' and 'Price_euros')
boolean_columns = ['Touchscreen', 'Ips', 'Company_Acer', 'Company_Apple', 
                   'Company_Asus', 'Company_Dell', 'Company_HP', 
                   'OpSys_Other', 'OpSys_Windows', 'Cpu_name_AMD', 
                   'Cpu_name_Intel Core i3', 'Cpu_name_Intel Core i5', 
                   'Cpu_name_Intel Core i7', 'Cpu_name_Other', 
                   'Gpu_name_AMD', 'Gpu_name_Intel', 'Gpu_name_Nvidia']

# Convert specified boolean columns to integers
data[boolean_columns] = data[boolean_columns].astype(int)
```

```python
data.head(2)
```

```python
x = data.drop('Price_euros',axis=1) # x values
y = data['Price_euros']                    
```

```python
!pip install scikit-learn
                    
  
```

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.25)
```

```python
x_train.shape,x_test.shape
```

```python
def model_acc(model):
    model.fit(x_train, y_train)
    acc = model.score(x_test, y_test)
    print(str(model) + " --> " + str(acc))
```

```python
from sklearn.linear_model import LinearRegression
lr = LinearRegression()
model_acc(lr)
```

```python
from sklearn.linear_model import LinearRegression
from sklearn.linear_model import Lasso
from sklearn.tree import DecisionTreeRegressor
from sklearn.ensemble import RandomForestRegressor
            
# Instantiate models
lr = LinearRegression()
model_acc(lr)

lasso = Lasso()
model_acc(lasso)

dt = DecisionTreeRegressor()
model_acc(dt)

rf = RandomForestRegressor()
model_acc(rf)
```

```python
#let choose random forest then use hypoparameters
from sklearn.model_selection import GridSearchCV

# Define the parameter grid
parameters = {
    'n_estimators': [10, 50, 100],
    'criterion': ['squared_error', 'absolute_error', 'poisson']
}

# Create a GridSearchCV object
grid_obj = GridSearchCV(estimator=rf, param_grid=parameters)

# Fit the grid search to the training data
grid_fit = grid_obj.fit(x_train, y_train)

# Get the best model
best_model = grid_fit.best_estimator_

# Output the best model
best_model
```

```python
best_model.score(x_test,y_test)
```

```python
# save model
import pickle

# Save the best model to a file
with open('predictor.pickle', 'wb') as file: #wb is the mood that open
    pickle.dump(best_model, file)
                                        
```
